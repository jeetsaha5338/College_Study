```python
import pandas as pd
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt
import numpy as np
df=pd.read_csv("binary.csv")
print(df.head())
```

```python
df.columns = ["admit","gre","gpa","prestige"]
print(df.columns)
```

```python
print(df.describe())
```

```python
print(df.prestige.unique())
```

```python
dummy_ranks=pd.get_dummies(df['prestige'],prefix="prestige")
print(type(dummy_ranks))
dummy_ranks
```

```python
cols_to_keep=['admit','gre','gpa']
data=df[cols_to_keep].join(dummy_ranks.loc[:,'prestige_2':])
data.head()
```

```python
data_tr=data.loc[0:250]
data_ta=data.loc[251:]
train_cols= data.columns[1:]
logit= smf.Logit(data_tr['admit'],data_tr[train_cols])
result=logit.fit()          
```

```python
print(result.params)
print(result.summary())
```

```python
result.conf_int()#bujhini
```

```python
result.pred_table()
```

```python
data_ta['predition']=result.predict(data_ta[train_cols])
data_ta.head()
```

```python
data_ta['admit_pred']=data_ta['predition']>0.5
data_ta
```
