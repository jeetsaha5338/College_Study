```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
data=pd.read_csv('http://www-bcf.usc.edu/~gareth/ISL/Advertising.csv',index_col=0)#FISRT ER TA INDEX HISABE NEOAR JNNE
data
data.shape
```

```python
fig, axs=plt.subplots(1,3,sharey=False)  #SHARING Y AXIS TO ALL SUBPLOTS....#True korle same rows index share hbe
data.plot(kind='scatter',x='TV',y='sales',ax=axs[0],figsize=(16,8))
data.plot(kind='scatter',x='radio',y='sales',ax=axs[1])
data.plot(kind='scatter',x='newspaper',y='sales',ax=axs[2])
```

```python
import statsmodels.formula.api as smf
lm = smf.ols(formula='sales~TV', data=data).fit()  #ordinary list square(ols) #fit ta regerssion kore TRAINING DATA
lm.params #print the coeffiecients
```

```python
import pandas as pd
x=pd.DataFrame({'TV': [125,220,298]})
x

```

```python
PREDICT=lm.predict(x)
```

```python
data.plot(kind='scatter',x='TV',y='sales')

plt.plot(x,PREDICT,c='red',linewidth=3)
```
