```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf
%matplotlib inline
data=pd.read_csv('servo.csv',header=None,names=['ONE','TWO','THREE','FOUR','FIVE'])#FISRT ER TA INDEX HISABE NEOAR JNNE
data['A_1']=data.ONE.map({'A':1,'B':0,'C':0,'D':0,'E':0})
data['B_1']=data.ONE.map({'A':0,'B':1,'C':0,'D':0,'E':0})
data['C_1']=data.ONE.map({'A':0,'B':0,'C':1,'D':0,'E':0})
data['D_1']=data.ONE.map({'A':0,'B':0,'C':0,'D':1,'E':0})
data['E_1']=data.ONE.map({'A':0,'B':0,'C':0,'D':0,'E':1})

data['A_2']=data.ONE.map({'A':1,'B':0,'C':0,'D':0,'E':0})
data['B_2']=data.TWO.map({'A':0,'B':1,'C':0,'D':0,'E':0})
data['C_2']=data.TWO.map({'A':0,'B':0,'C':1,'D':0,'E':0})
data['D_2']=data.TWO.map({'A':0,'B':0,'C':0,'D':1,'E':0})
data['E_2']=data.TWO.map({'A':0,'B':0,'C':0,'D':0,'E':1})
data.head()
data_train=data[:120]
data_test=data[120:]
print(data_train.shape)
print(data_train)
print(data_test.shape)
print(data_test)


print(data)
print('===================================')


```

```python
lm = smf.ols(formula='FIVE~THREE+FOUR+A_1+B_1+C_1+D_1+E_1+A_2+B_2+C_2+D_2+E_2', data=data_train).fit()  #ordinary list square(ols) #fit ta regerssion kore TRAINING DATA
print(lm.params)#print the coeffiecients
print(lm.summary())
```

```python
x=data_test[['THREE','FOUR','A_1','B_1','C_1','D_1','E_1','A_2','B_2','C_2','D_2','E_2']]
#x=data_test.ix[:,['TV','radio']]#IT WILL ALSO WORK
y_pre=lm.predict(x)
print("===================================================")
print(x)
y_ac=np.array(data_test['FIVE'])#ACTUAL CELLS
y_np=np.array(y_pre)#PREDICTED CELLS
print("===================================================")
print(y_ac)
print("===================================================")
print(y_np)
er1=(y_ac-y_np)**2
er1_mean=np.mean(er1)
er1_sqrt=np.sqrt(er1_mean)
print("===================================================")
print(er1_sqrt)
```
